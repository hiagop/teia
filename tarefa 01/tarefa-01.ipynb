## 1. Baseline  

Considerando os modelos matemáticos que representam o crescimento de dada população, para um estudo onde se quer identificar quais modificações, ou mutações, mais contribuem para a sobrevivência dos indivíduos desta população, seria interessante tomar como base para comparação o modelo em que a população cresce de maneira mais contida, sem que haja incrementos demasiadamente grandes.
Como pode ser visto a seguir, dois modelos bastante comuns são comparados: o logístico $$f(x)=\frac{1}{1-\exp(-x)}$$ e o exponencial $$g(x)=\exp(x)$$

```python
import matplotlib.pyplot as pl
import numpy as np
import pandas as pd
import math
```

```python
yl, ye = [], []
for i in range(0, 26):
    yl.append(1/1-math.pow(1.2,-i))
    ye.append(0.11*math.pow(1.2,i)/10)

yl, ye = np.array(yl), np.array(ye)   

pl.figure(figsize=(10, 5))
pl.title('Curvas de Crescimento')
pl.xlabel('X')
pl.ylabel('Y')
pl.plot(yl, label='Logística', color='orange')
pl.plot(ye, label='Exponencial', color='blue')
pl.legend(loc='best')
pl.show()
```

Dos dois modelos comparados, o que melhor representa um crescimento ideal de uma população é o logístico. Portanto, todas as _baselines_ utilizadas terão um comportamento similar ao da curva logística exibida na figura acima.   

As _baselines_ são obtidas combinando as opções de habitats e fatores de seleção: região ártica e equatorial, e seleção por escassez de comida e por predadores; e as mutações que geram melhor resposta para dada combinação de ambiente. Sendo assim, temos as seguintes _baselines_:
 * HA + FC + MD
 * HE + FC + MD
 * HA + FP + MC
 * HE + FP + MP
 
onde  

HA = HABITAT ÁRTICO  
HE = HABITAT EQUATORIAL  
FC = ESCASSEZ DE COMIDA  
FP = PREDADORES  
MC = COMPRIMENTO DA CAUDA  
MD = TAMANHO DOS DENTES  
MP = COR DOS PELOS  
MN = SEM MUTAÇÃO  

__Obs.: As combinações que não tomavam nenhum fator seletivo foram desconsideradas, pois isso causaria um crescimento descontrolado da população, assumindo um comportamento exponencial.__  

Abaixo observamos os experimentos utilizados como base para comparação com os demais. Vale ressaltar que as combinações escolhidas para se usar como _baseline_ são aquelas em que a mutação contribui para a sobrevivência da população mutada, ou que, pelo menos, não compromete a prevalência da população como um todo.

```python
ha_fc_md = pd.read_csv("exp/ha.fc.md.csv", sep=',', index_col=0)

data = ha_fc_md.values.astype('int')

t1, dc1, dl1 = [], [], []
for row in data:
    t1.append(row[0])
    dc1.append(row[1])
    dl1.append(row[2])

t1, dc1, dl1 = np.array(t1), np.array(dc1), np.array(dl1)

pl.figure(figsize=(10, 5))
pl.title('HA + FC + MD')
pl.xlabel('Gerações')
pl.ylabel('Coelhos')
pl.plot(t1, label='Total de Coelhos', color='black')
pl.plot(dc1, label='Dentes Curtos', color='orange')
pl.plot(dl1, label='Dentes Longos', color='blue')
pl.legend(loc='best')
pl.show()
```

Nas duas primeiras _baselines_, observa-se como a aparição de indivíduos com dentes mais longos afeta o desenvolvimento da população, idenpendentemente de qual o habitat analisado. Este comportamento se deve a maior facilidade adquirida por estes indivíduos em obter alimento.

```python
he_fc_md = pd.read_csv("exp/he.fc.md.csv", sep=',', index_col=0)

data = he_fc_md.values.astype('int')

t5, dc2, dl2 = [], [], []
for row in data:
    t5.append(row[0])
    dc2.append(row[1])
    dl2.append(row[2])

t5, dc2, dl2 = np.array(t5), np.array(dc2), np.array(dl2)

pl.figure(figsize=(10, 5))
pl.title('HE + FC + MD')
pl.xlabel('Gerações')
pl.ylabel('Coelhos')
pl.plot(t5, label='Total de Coelhos', color='black')
pl.plot(dc2, label='Dentes Curtos', color='orange')
pl.plot(dl2, label='Dentes Longos', color='blue')
pl.legend(loc='best')
pl.show()
```

## 3. Hipóteses  

Analisando os resultados de cada teste, se torna possível gerar hipóteses a respeito do impacto de cada mutação nos diferentes cenários estudados. Estas hipóteses serão verificadas na [seção seguinte](#4.-Resultados), onde o crescimento de cada população de indivíduos mutados é comparado com os demais. As hipóteses a seguir levam em consideração somente os dados obtidos através dos experimentos, não tomando como base fatores externos aos possíveis de se obter por meio do simulador.

1. Frente a fatores seletivos mais rígidos, apenas uma das modificações podem ser benéficas à população, não dando brechas a indivíduos menos capazes de se adaptar ao ambiente. Por exemplo, quando tratamos da disputa por alimento, independente do ambiente, somente aqueles que conseguem obter uma quantidade maior de alimento (os de dentes mais longos) prevalecerão, restando aos demais a extinção.
2. Quando a seleção é feita por meio da predação, o habitat também influencia no crescimento das populações.
3. Mutações benéficas para um cenário podem ser consideradas neutras para os demais. Enquanto que mutações neutras em outro cenário podem ser prejudiciais nos restantes.


## 4. Resultados  

Avaliando as curvas de crescimento populacional comparadas entre si, nos quatro cenários propostos, podemos validar as hipóteses geradas na [seção anterior](#3.-Hipóteses). 

Nos dois primeiros gráficos mostrados a seguir, validamos a hipótese de que o ambiente não influencia o crescimento da população, quando se leva em consideração que a única frente de controle é a escassez de alimentos, e que o único meio de sobrevicência é se adaptar a quantidade de recursos disponíveis. Portanto, apenas uma das mutações se torna benéfica para a população.


```python
pl.figure(figsize=(10, 5))
pl.title('HA + FC')
pl.xlabel('Gerações')
pl.ylabel('Coelhos')
pl.plot(dl1, label='Dentes Longos*', color='black')
pl.plot(cl1, label='Cauda Longa', color='blue', alpha=0.5)
pl.plot(pm1, label='Pelo Marrom', color='orange', alpha=0.5)
pl.plot(t4, label='Sem Mutação', color='red', alpha=0.5)
pl.legend(loc='best')
pl.show()
```

```python
pl.figure(figsize=(10, 5))
pl.title('HE + FC')
pl.xlabel('Gerações')
pl.ylabel('Coelhos')
pl.plot(dl2, label='Dentes Longos*', color='black')
pl.plot(cl2, label='Cauda Longa', color='blue', alpha=0.5)
pl.plot(pm2, label='Pelo Marrom', color='orange', alpha=0.5)
pl.plot(t8, label='Sem Mutação', color='red', alpha=0.5)
pl.legend(loc='best')
pl.show()
```

Com os próximos dois gráficos, também confirma-se a hipótese de que o ambiente se torna um fator a se considerar quando a pressão seletiva vem dos predadores. Já que os indivíduos cuja mutação modifica a coloração dos seus pelos são rapidamente excluídos do meio, quando o ambiente é o Ártico; mas se tornam os mais aptos à sobrevivência quando mudamos para a região equatorial. 

```python
pl.figure(figsize=(10, 5))
pl.title('HA + FP')
pl.xlabel('Gerações')
pl.ylabel('Coelhos')
pl.plot(cl3, label='Cauda Longa*', color='black')
pl.plot(dl3, label='Dentes Longos', color='blue', alpha=0.5)
pl.plot(pm3, label='Pelo Marrom', color='orange', alpha=0.5)
pl.plot(t12, label='Sem Mutação', color='red', alpha=0.5)
pl.legend(loc='best')
pl.show()
```

Por fim, confirmamos que não há uma mutação chave, que sirva para todo e qualquer ambiente. De modo que mutações tidas como proveitosas em certos cenários podem não afetar o crescimento daquela população em outro; ou ainda, prejudicar seu desenvolvimento nos demais. Por exemplo, vemos em todos os gráficos apresentados, que a mutação mais impactante no clima ártico (comprimento dos dentes) também tem o mesmo impacto na região equatorial, mas somente quando a seleção é feita por meio da disputa por alimento. Nos demais cenários, esta mutação não tem impacto algum, permitindo que ambos os grupos sobrevivam, ou não surtem o efeito desejado para a sobrevivência da espécie.

```python
pl.figure(figsize=(10, 5))
pl.title('HE + FP')
pl.xlabel('Gerações')
pl.ylabel('Coelhos')
pl.plot(pm4, label='Pelo Marrom*', color='black')
pl.plot(cl4, label='Cauda Longa', color='blue', alpha=0.5)
pl.plot(dl4, label='Dentes Longos', color='orange', alpha=0.5)
pl.plot(t16, label='Sem Mutação', color='red', alpha=0.5)
pl.legend(loc='best')
pl.show()
```
